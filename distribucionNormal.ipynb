```python
import numpy as np
from numpy.random import binomial
from scipy.stats import binom
from math import factorial
import matplotlib.pyplot as plt
```

Escribiendo en python secuencia de eventos tipo Bernoulli

```python
def binomial_fuc(k, n, p):
    return (factorial(n)/(factorial(k)*factorial(n-k)))*(p**k*((1-p)**(n-k)))
```

```python
binomial_fuc(2,3,0.5)
```

# Simulaciones de secuencias con generadores aleatorios.

- Los generadores aleatorios tienen como proposito simular muestras de datos que resultarian de muestreos en la vida real de procesos aleatorios como lanzar una moneda o un dado.

```python
# simulacion con numpy random
p = 0.5
n = 3
binomial(n,p)
```

```python
#Lanzar una moneda con un for que se recorra 100 veces.

arr = []
for _ in range(100):
    arr.append(binomial(n,p))
print(arr)
#Probabilidades de cada numero
np.unique(arr, return_counts = True)[1]/len(arr)
```

```python
def plot_hist(num_trail):
    values = [0,1,2,3]
    arr = []
    for _ in range(num_trail):
        arr.append(binomial(n,p))
    sim = np.unique(arr, return_counts = True)[1]/len(arr)
    #Distribucion teorica.
    teorica = [binom(3, 0.5).pmf(k)  for k in values]
    plt.bar(values,sim, color = 'green')
    plt.bar(values,teorica, alpha = 0.3, color = 'blue')
    plt.title(f'{num_trail} experimentos')
    plt.show()
   
       
plot_hist(1000000)
        
```
