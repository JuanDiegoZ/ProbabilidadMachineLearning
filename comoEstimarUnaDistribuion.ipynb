```python
import numpy as np
from matplotlib import pyplot as plt
from numpy.random import normal
from scipy.stats import norm
```

```python
#Generar datos artificiales.
sample = normal(size = 10000)
pyplot.hist(sample,bins = 60)
pyplot.show()
```

## Estimacion parametrica.


```python
#Creando datos artificiales con mu = 50, sigma = 5
sample = normal(loc = 50, scale=5, size = 100000)
mu = sample.mean()
sigma = sample.std()
dist = norm(mu,sigma)
value = [values for values in range (30,70)]
probabilidades = [dist.pdf(values) for values in value]
plt.hist(sample,bins=100,density=True)
plt.plot(value, probabilidades)
plt.show()
```

## Estimacion no parametrica

```python
from numpy import hstack
from sklearn.neighbors import KernelDensity

#construimos una distribucin bimodal
sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size=700)
sample = hstack((sample1, sample2))

model = KernelDensity(bandwidth=2, kernel='gaussian')
sample = sample.reshape((len(sample), 1))
model.fit(sample)

values = np.asarray([value for value in range(1, 60)])
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values) #probabilidad logartmica
probabilities = np.exp(probabilities)  # inversin de probabilidad

pyplot.hist(sample, bins=50, density=True) 
pyplot.plot(values[:], probabilities)
pyplot.show()
```
